```python
from keras.models import load_model
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing import image
```

```python
img = image.load_img('D:/Work Prof/Task 18 dr Ria Nova/03 Klasifikasi/Uji Kesesuaian 2/all new/ALULA_RIZKI_4CH_20.jpg', grayscale=False, target_size=(64, 64))
show_img=image.load_img('D:/Work Prof/Task 18 dr Ria Nova/03 Klasifikasi/Uji Kesesuaian 2/all new/ALULA_RIZKI_4CH_20.jpg', grayscale=False, target_size=(200, 200))
disease_class=['normal','abnormal']
x = image.img_to_array(img)
x = np.expand_dims(x, axis = 0)
x /= 255

custom = model.predict(x)
print(custom[0])

plt.imshow(show_img)
plt.show()

a=custom[0]
ind=np.argmax(a)
print('Prediksi')
print('Prediction:',disease_class[ind])
print()

```
